```python
# 그림 그려주는 라이브러리 import

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# 'Cicle'이라는 class를 정의

class Circle(object):

    # 기본값 정의
    def __init__(self, radius, color='blue'):
        self.radius = radius
        self.color = color

    # r값을 추가하는 방법 정의
    def add_radius(self, r):
        self.radius = self.radius + r
        return(self.radius)

    # Circle그리는 방법 정의
    def drawCircle(self):
        plt.gca().add_patch(plt.Circle((0, 0), radius=self.radius, fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python
GreenCircle = Circle(0, 'green') # r=0, color=green인 'GreenCircle' 생성
GreenCircle.add_radius(2) # r+2
GreenCircle.add_radius(2) # r+2
GreenCircle.radius
```

```python
  class Rectangle(object):

    # Rectangle이라는 class의 변수를 정의해볼께
    def __init__(self, width = 2, height = 3, color ='r'):
        self.height = height
        self.width = width
        self.color = color

    # Method: Ractangle을 그려주는 방법
    def drawRectangle(self):
        plt.gca().add_patch(plt.Rectangle((0, 0), self.width, self.height ,fc=self.color))
        plt.axis('scaled')
        plt.show()
```

```python
SkinnyBlueRectangle = Rectangle(2, 10, 'blue')
print(SkinnyBlueRectangle.height)
print(SkinnyBlueRectangle.width)
print(SkinnyBlueRectangle.color)
SkinnyBlueRectangle.drawRectangle()
```

```python
FatYellowRectangle = Rectangle(20, 5, 'yellow')
print(FatYellowRectangle.height)
print(FatYellowRectangle.width)
print(FatYellowRectangle.color)
FatYellowRectangle.drawRectangle()
```
