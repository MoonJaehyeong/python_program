```python
#Q7. 개별 분석: (도서) 주제분류별 20년, 23년 도서권수를 비교, 분석
# 로직 순서
# 1)주제분류번호를 그룹핑하기 위한 'tag'열을 추가
# 2)그룹핑을 위해 주제분류열의 값(기존: 문자열) 앞 3자리를 추출하여 새로운 'topic'열에 저장
# 3)추출한 3자리의 노이즈를 제거하고 정수값으로 변환
# 4)그룹핑을 위한 그룹 함수 정의
# 5) 그룹 함수를 기준으로 topic(정수)값을 판단하여 'tag'열에 그룹 함수에서 정의된 그룹 고유번호를 부여
# 6) 20년, 23년 각각 수행
# 7) 각각 그룹 고유번호(도서주제분류)별로 도서권수 총합을 산출하여 'tag'열에 저장
# 8) 도서주제분류별로 20년, 23년 도서권수 df를 merge
# 9) 도서주제분류별 20년, 23년 도서권수를 비교하기 위한 그래프 그리기
```

```python
book_20y_df['tag'] = 0 # 20년, 23년 각각 'tag'열 추가
book_23y_df['tag'] = 0
```

```python
book_20y_df['topic'] = 0 # 20년, 23년 각각 'topic'열 추가
book_23y_df['topic'] = 0
```

```python
df_2020 = book_20y_df[['주제분류번호', 'topic', '도서권수', '등록일자', 'tag']]
df_2023 = book_23y_df[['주제분류번호', 'topic', '도서권수', '등록일자', 'tag']]
```

```python
df_2020['topic'] = df_2020['주제분류번호'].str[:3] #그룹핑을 위해 주제분류열의 값(기존: 문자열) 앞 3자리를 추출하여 새로운 'topic'열에 저장
```

```python
df_2023['topic'] = df_2023['주제분류번호'].str[:3] #그룹핑을 위해 주제분류열의 값(기존: 문자열) 앞 3자리를 추출하여 새로운 'topic'열에 저장
```

```python
df_2020.head()
```

```python
df_2023.head()
```

```python
df_2020['topic'] = df_2020['topic'].fillna(0).astype(int) #추출한 3자리의 노이즈를 제거하고 정수값으로 변환
```

```python
df_2023['topic'] = df_2023['topic'].fillna(0).astype(int) #추출한 3자리의 노이즈를 제거하고 정수값으로 변환
```

```python
def find_group(value): 
    groups = [100, 200, 300, 400, 500, 600, 700, 800, 900, 1000]
    for i in range(len(groups)):
        if value < groups[i]:
            if i == 100:
                return 100
            
            return groups[i-1] #그룹핑을 위한 그룹 함수 정의
```

```python
df_2020['tag'] = [find_group(e) for e in df_2020.topic] #그룹 함수를 기준으로 topic(정수)값을 판단하여 'tag'열에 그룹 함수에서 정의된 그룹 고유번호를 부여
```

```python
df_2023['tag'] = [find_group(e) for e in df_2023.topic] #그룹 함수를 기준으로 topic(정수)값을 판단하여 'tag'열에 그룹 함수에서 정의된 그룹 고유번호를 부여
```

```python
df_2020.head()
```

```python
df_2023.head()
```

```python
tag_groups_2020 = df_2020.groupby('tag') #tag(도서주제분류)가 같은 그룹으로 묶음
```

```python
tag_groups_2023 = df_2023.groupby('tag') #tag(도서주제분류)가 같은 그룹으로 묶음
```

```python
tag_sum_2020_df = tag_groups_2020.sum() #tag 그룹별 합계 즉, 2020년 tag 그룹별 총 도서권수 계산
```

```python
tag_sum_2023_df = tag_groups_2023.sum() #tag 그룹별 합계 즉, 2023년 tag 그룹별 총 도서권수 계산
```

```python
tag_sum_2020_df.reset_index(inplace=True) #기준 인덱스 수정
```

```python
tag_sum_2023_df.reset_index(inplace=True) #기준 인덱스 수정
```

```python
tag_sum_2020_df.head()
```

```python
tag_sum_2023_df.head()
```

```python
tag_sum_df = pd.merge(tag_sum_2020_df, tag_sum_2023_df, on='tag', how='inner', suffixes=('_2020', '_2023'))
#tag를 기준으로 두개의 df를 합침
```

```python
tag_sum_df_1 = tag_sum_df[['tag' ,'도서권수_2020', '도서권수_2023']] #tag(도서 주제 분류)별 도서권수
```

```python
tag_sum_df_1.head()
```

```python
# tag(도서 주제 분류)별 20년, 23년 도서권수를 비교하기 위한 그래프 그리기
```
